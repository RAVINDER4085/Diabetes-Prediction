Importing the Dependencies


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from  sklearn.model_selection import train_test_split
from sklearn.svm import SVC
```

Data Collection and Analysis

PIMA Diabetes Dataset

```python
df=pd.read_csv('/content/diabetes.csv')
df.head()
```

```python
X = df.drop(columns = 'Outcome', axis=1)
Y = df['Outcome']
```

```python
print(X)
```

```python
# getting the statistical measures of the data
df.describe()
```

0 --> Non-Diabetic

1 --> Diabetic

```python
df.groupby('Outcome').mean()
```

Data Standardization

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X = sc.fit_transform(X)
```

```python
print(X)
```

Spliting the dataset into training and test dataset

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,Y,random_state=0,test_size=0.25)
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
print(x_train)
```

Applying the support vector classifier model

```python
from sklearn.svm import SVC
classifier=SVC(kernel='linear',random_state=0)
classifier.fit(x_train,y_train)
```

Predicting for training dataset

```python
x_train_prediction=classifier.predict(x_train)
```

Model Evaluation
Accuracy Score

```python
from sklearn.metrics import accuracy_score,confusion_matrix
cm=confusion_matrix(y_train,x_train_prediction)
print(cm)
accuracy_score(y_train,x_train_prediction)
```

Predicting for test dataset

```python
x_test_prediction=classifier.predict(x_test)
```

Accuracy score for test dataset

```python
cm=confusion_matrix(y_test,x_test_prediction)
print(cm)
accuracy_score(x_test_prediction,y_test)
```

```python
#Making a predictive system
input_data=(8,183,64,0,0,23.3,0.672,32)

#changing the input_data to numpy array
numpy_array=np.asarray(input_data)

## reshape the array as we are predicting for one instance
numpy_array_reshape=numpy_array.reshape(1,-1)

#standardised the input data
std_data=sc.transform(numpy_array_reshape)
print(std_data)

#predict the input 
prediction=classifier.predict(std_data)
print(prediction)

if (prediction[0] == 0):
  print('The person is not diabetic')
else:
  print('The person is diabetic')
```
